```python
N = [32,64,128,256]

fig = plt.figure(figsize=(10, 6.5))
ax = fig.add_subplot(111)

Ns = []
err = []
for n in N:
    #"Ising/ForPhase/properties_is_0_higher_hpc.txt"
     
    filename = "CSR_float_"+str(n)+".log"
    
    with open(filename) as f:
        my_lines = f.readlines()[1].split()[-1]#.replace('\n', '')
    Ns.append(n)
    err.append(float(my_lines))
    print(my_lines)
plt.loglog(Ns, err, "o--", label = "float, CSR")

Ns = []
err = []
for n in N:
    #"Ising/ForPhase/properties_is_0_higher_hpc.txt"
     
    filename = "CSR_double_"+str(n)+".log"
    
    with open(filename) as f:
        my_lines = f.readlines()[1].split()[-1]#.replace('\n', '')
    Ns.append(n)
    err.append(float(my_lines))
    print(my_lines)
plt.loglog(Ns, err, "o--", label = "double, CSR")

Ns = []
err = []
for n in N:
    #"Ising/ForPhase/properties_is_0_higher_hpc.txt"
     
    filename = "ELL_float_"+str(n)+".log"
    
    with open(filename) as f:
        my_lines = f.readlines()[1].split()[-1]#.replace('\n', '')
    Ns.append(n)
    err.append(float(my_lines))
    print(my_lines)
plt.loglog(Ns, err, "o--", label = "float, ELL-S")

Ns = []
err = []
for n in N:
    #"Ising/ForPhase/properties_is_0_higher_hpc.txt"
     
    filename = "ELL_double_"+str(n)+".log"
    
    with open(filename) as f:
        my_lines = f.readlines()[1].split()[-1]#.replace('\n', '')
    Ns.append(n)
    err.append(float(my_lines))
    print(my_lines)
plt.loglog(Ns, err, "o--", label = "double, ELL-S")

plt.xlabel(r'N', fontsize = 22)
plt.ylabel(r"Discretization error ", fontsize=22)
plt.tick_params(axis='x', labelsize=22) 
plt.tick_params(axis='y', labelsize=22) 

plt.legend(loc="best", fontsize=18)
plt.savefig("Report/errors.png")
```
