```python
fig = plt.figure(figsize=(13, 10))
ax = fig.add_subplot(111)

data = pd.read_csv("Theory/uppmax_02.txt", sep='\s+', header=None)
plt.loglog(data[4], data[15], "s",label ="02, uppmax")

data = pd.read_csv("Theory/uppmax_03.txt", sep='\s+', header=None)
plt.loglog(data[4], data[15], "s",label ="03, uppmax")

data = pd.read_csv("Theory/my_03.txt", sep='\s+', header=None)
plt.loglog(data[4], data[15], "*",label ="03, GenuineIntel")

data = pd.read_csv("Theory/my_simd.txt", sep='\s+', header=None)
plt.loglog(data[4], data[15], "*",label ="SIMD, GenuineIntel")

data = pd.read_csv("Theory/uppmax_simd.txt", sep='\s+', header=None)
plt.loglog(data[4], data[15], "s",label ="SIMD, uppmax")

plt.title("Memory bandwidth:  memory operations per second", fontsize=24)
plt.ylabel("GB/s", fontsize=20)

plt.tick_params(axis='x', labelsize=20) 
plt.tick_params(axis='y', labelsize=20) 

plt.legend(fontsize=15)

plt.savefig("uppmax_memory.png", bbox_inches='tight')
```
